```python

import numpy as np

class Perceptron:
    def __init__(self, weights, bias):
        # Convierte los pesos a un array de numpy para facilitar los cálculos
        self.weights = np.array(weights)
        self.bias = bias

    def step_function(self, z):
        # La función de activación escalón (heaviside)
        return 1 if z >= 0 else 0

    def predict(self, inputs):
        # Convierte las entradas a un array de numpy
        inputs = np.array(inputs)

        # Cálculo del "puntaje z" (suma ponderada + bias)
        # Esto también se conoce como la entrada neta o weighted sum
        z = np.dot(inputs, self.weights) + self.bias

        # Decisión de salida (clasificación) aplicando la función de activación
        output = self.step_function(z)

        return z, output

print("Clase Perceptron definida.")
```

```python
# Definir los pesos y el bias iniciales para el perceptrón
# Estos valores se pueden ajustar para ver cómo cambia la clasificación
weights = [0.5, -0.2] # Ejemplo: peso para feature_1 y feature_2
bias = -0.1

# Instanciar el perceptrón
perceptron = Perceptron(weights, bias)

# Conjunto de pruebas (casos controlados)
# Cada fila es un caso de prueba con dos características [feature_1, feature_2]
test_cases = [
    [0.1, 0.2],   # Caso 1
    [0.8, 0.1],   # Caso 2
    [-0.5, 0.6],  # Caso 3
    [0.3, -0.4],  # Caso 4
    [1.0, 1.0],   # Caso 5
    [-1.0, -1.0]  # Caso 6
]

print(f"Perceptrón inicializado con pesos: {weights} y bias: {bias}")
print(f"Casos de prueba definidos: {test_cases}")
```

### Resultados de la Clasificación del Perceptrón

```python
print(f"\n{'='*50}")
print(f"Parámetros del Perceptrón: Pesos = {perceptron.weights}, Bias = {perceptron.bias}")
print(f"{'='*50}")

# Imprimir encabezado de la tabla
print(f"{'Entradas':<15} | {'Pesos':<15} | {'Bias':<8} | {'Puntaje Z (Suma Ponderada)':<30} | {'Decisión de Salida':<20}")
print(f"{'-'*15} | {'-'*15} | {'-'*8} | {'-'*30} | {'-'*20}")

# Ejecutar pruebas y mostrar resultados
for i, inputs in enumerate(test_cases):
    # Para cada caso de prueba, se predice la salida del perceptrón
    # y se imprimen las entradas, pesos, bias, el puntaje Z y la decisión final.
    z_score, output_decision = perceptron.predict(inputs)
    print(f"Caso {i+1}: {str(inputs):<10} | {str(list(perceptron.weights)):<15} | {perceptron.bias:<8.2f} | {z_score:<30.4f} | {output_decision:<20}")

print(f"{'='*50}")
```

### Análisis de los Parámetros (Pesos y Bias) en la Decisión Final

En un perceptrón, la decisión de clasificación se toma en base a si la **suma ponderada (puntaje Z)** de las entradas más el bias es mayor o igual a cero. Si `Z >= 0`, la salida es 1; de lo contrario, es 0.

-   **Pesos (`weights`):**
    -   Los **pesos** indican la importancia de cada característica de entrada.
    -   **Peso positivo grande**: esa característica empuja el puntaje Z hacia arriba, aumentando la probabilidad de que la salida sea 1.
    -    **Peso negativo grande** esa característica empuja el puntaje Z hacia abajo, aumentando la probabilidad de que la salida sea 0.
    - Al cambiar los pesos, se modifica la inclinación u orientación de la frontera de decisión (el hiperplano que separa las clases).
    - Ejemplo: si aumentamos el peso de una característica, el perceptrón se vuelve más sensible a ella.




-   **Bias (`bias`):**
    - El **bias** funciona como un umbral o un desplazamiento.
    - Permite mover la frontera de decisión sin cambiar su orientación.
    - **Bias positivo**: el perceptrón tiende a activarse más fácilmente, incluso si las entradas son pequeñas o cero.
    - **Bias negativo**: el perceptrón necesita un puntaje Z más alto para activarse, lo que lo hace más “exigente”.


**En conclusión:**

- Los **pesos** definen la dirección y sensibilidad hacia las características de entrada.
- El **bias** ajusta el punto de corte o la tendencia general del perceptrón a activarse.
- Juntos, pesos y bias determinan la frontera de decisión que permite clasificar datos en diferentes categorías.
- Una correcta **sintonización** de estos parámetros es esencial para que el perceptrón aprenda y clasifique de manera efectiva
